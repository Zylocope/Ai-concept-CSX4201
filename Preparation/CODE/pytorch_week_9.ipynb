```python
# Assignment1
import torch
X1 = torch.rand(5,6)
print(X1)
print(X1.shape) # shape of X1
print(X1.dtype) # data type of X1
print(torch.mean(X1)) # mean of all elemnets in X1
```

```python
# Assignment2
X2 = torch.randint(0, 11, (4,4), dtype=torch.int)
print(X2)
X2_float = X2.float()
print(X2_float)
print(X2_float - X2)
'''
The difference between x2_float and x2 is zero because converting the tensor 
from integer to float does not change the values of the elements. The only change 
is in the data type of the elements. Therefore, when we subtract the original integer 
tensor from the converted float tensor, the result is a tensor of zeros.
'''
```

```python
# Assignment3
X3 = torch.rand(3,2)
print(X3)

X3_transposed = torch.transpose(X3, 0, 1)
print(X3_transposed)

matrix_multi = X3 @ X3_transposed
print(matrix_multi)
print(matrix_multi.dtype)
```

```python
# Assignment4
X4_A = torch.randint(1, 5, (2,3), dtype=torch.int)
X4_B = torch.randint(1, 5, (2,3), dtype=torch.int)

add = X4_A + X4_B
print("add: ", add)
print("add: ", add.shape)

sub = X4_A - X4_B
print("sub: ", sub)
print("sub: ", sub.shape)

multi = X4_A * X4_B
print("multi: ", multi)
print("multi: ", multi.shape)
```

```python
# Assignment6
X6 = torch.rand(4,4)
print(X6)

# Calculate row sum
row_sum = X6.sum(dim=1)
print("row_sum: ", row_sum)

# Calculate column sum
col_sum = X6.sum(dim=0)
print("col_sum: ", col_sum)
```

```python
# Assignment7
X7_A = torch.rand(2,2)
X7_B = torch.rand(2,2)

# Concatenate them along dimension 0
result_di0 = torch.cat((X7_A, X7_B), dim=0) # result shape (4,2)

# Concatenate them along dimension 1
result_di1 = torch.cat((X7_A, X7_B), dim=1) # result shape (2,4)

print("X7_A:", X7_A)
print("X7_B:", X7_B)
print("result_di0", result_di0)
print("result_di1", result_di1)
```

```python
# Assignment9
X9 = torch.rand(3,3)
determinant = torch.linalg.det(X9)
print(determinant)

'''
The determinant of a matrix provides information about the matrix's properties. 
If the determinant is zero or near zero, it indicates that the matrix is nearly 
singular or singular. A singular matrix does not have an inverse, and its rows or 
columns are linearly dependent.

If a determinant equals 0.0334 indicates that the matrix is nearly singular, and 
special care should be taken when using it in further calculations.
'''
```
